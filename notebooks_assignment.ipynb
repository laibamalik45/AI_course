```python
# Question 7:
mark1 = float(input("Enter marks for subject 1: "))
mark2 = float(input("Enter marks for subject 2: "))
mark3 = float(input("Enter marks for subject 3: "))
mark4 = float(input("Enter marks for subject 4: "))
mark5 = float(input("Enter marks for subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
percentage = (total / 500) * 100

print("Total marks:", total)
print("Percentage:", percentage)

if percentage >= 80:
    grade = 'A'
elif percentage >= 60:
    grade = 'B'
elif percentage >= 50:
    grade = 'C'
elif percentage >= 40:
    grade = 'D'
else:
    grade = 'F'

print("Grade:", grade)
```

```python
# Question 9 calculator:
input1 = int(input("enter your 1st value: "))
input2 = int(input("enter your 2nd value: "))
operand = input(("Enter your operand"))
if operand == "+":
    print("The value is: ",input1 + input2)
elif operand == "-":
    print("The value is: ",input1 - input2)
elif operand == "x":
    print("The value is: ",input1 * input2)
elif operand == "/":
    print("The value is: ",input1 / input2)
elif operand == "%":
    print("The value is: ",input1 % input2)
else:
    print("Please Enter correct operator")
```
